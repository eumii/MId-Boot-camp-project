```python
def sorting_age(x):
    if x <= 11:
        return 11
    elif x > 11 and x <= 24:
        return 24
    elif x > 24 and x <= 49:
        return 49
    elif x > 49 and x <= 69:
        return 69
    elif x >= 70:
        return 70
    else:
        return 0
```

# ANOVA testing

## Hypothesis 

****1.**** Defining null and alternative hypothesis. 

***H0*** : the coverage of the vaccination is the same for everyone in every region of France

***H1***: the alternative would be that the coverage IS NOT the same in every region of France.



**2.** Select your samples. You should choose samples that are representative of the larger population you are interested in and ensure that they are randomly chosen.

**3.** Calculate the means and standard deviations for each sample.

**4.** Calculate the overall mean for all samples combined.

**5.** Calculate the sum of squares for each sample and the total sum of squares.

**6.** Calculate the degrees of freedom for each sample and the total degrees of freedom.

**7.** Calculate the mean square for each sample and the overall mean square.

***H0*** : the coverage of the vaccination is the same for gender in every region of France

***H1***: the alternative would be that the coverage IS NOT the same for gender in every region of France.

```python
data_gender_count_test = data_gender_count_i.drop([0], axis = 0)
data_gender_count_test = data_gender_count_test.drop(['sexe'], axis = 1)
data_gender_count_test
```

```python
data_gender_count_test_list1 = data_gender_count_test.values[0].tolist()
print(data_gender_count_test_list1)
data_gender_count_test_list2 = data_gender_count_test.values[1].tolist()
print(data_gender_count_test_list2)
```

```python
import scipy.stats as stats

statistic, pvalue = stats.f_oneway(data_gender_count_test_list1, data_gender_count_test_list2)

print("F-statistic:", statistic)
print("p-value:", pvalue)
```

we fail to reject the H0, we can then says that the gender behave the same in the vaccination rates in the region of France.
